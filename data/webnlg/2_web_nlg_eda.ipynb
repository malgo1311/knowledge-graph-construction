```python
import os
import pandas as pd
import collections
```

```python
data_path = "/work/pi_dhruveshpate_umass_edu/project_19/Data/WebNGL"
file = "webnlg_dev.csv"

data = pd.read_csv(os.path.join(data_path, file))
data.head()
```

```python
print("*"*20)
print("Data shape")
print("*"*20)
print(data.shape, "\n")

print("*"*20)
print("Unique text comments")
print("*"*20)
print(data['comment'].value_counts(), "\n")

print("*"*20)
print("Unique triplet sizes")
print("*"*20)
print(data['size'].value_counts(), "\n")

print("*"*20)
print("Unique categories")
print("*"*20)
print(data['category'].value_counts(), "\n")
```

```python
categoryRelationMap = {}

for i in data.index:
    category = data['category'][i]
    if category not in categoryRelationMap:
        categoryRelationMap[category] = collections.defaultdict(int)
    
    triplets = eval(data['modified_triple_sets'][i])
    
    for item in triplets:
        relation = item.split(' | ')[1]
        categoryRelationMap[category][relation] += 1
    
    # break
```

```python
all_relations = []

print("Number of relations in each category")
for k, v in categoryRelationMap.items():
    print(f"{len(v.keys())} \t {k}")
    
print("")
for k, v in categoryRelationMap.items():
    print(f"{k} - {v.keys()}", "\n")
    all_relations.extend(list(v.keys()))
```

```python
all_relations = sorted(list(set(all_relations)))
len(all_relations)
```

```python
import numpy as np
from sklearn.cluster import AffinityPropagation
import distance
```

```python
words = np.asarray(all_relations) #So that indexing with a list will work
# words = all_relations
lev_similarity = -1*np.array([[distance.levenshtein(w1,w2) for w1 in words] for w2 in words])
```

```python
idx = 20
w1 = all_relations[idx]
w2 = all_relations[idx+1]
w1, w2, distance.levenshtein(w1,w2)
```

```python
type(lev_similarity), lev_similarity.shape, type(lev_similarity).__name__
```

```python
lev_similarity = np.array(lev_similarity, dtype=np.float64)
affprop = AffinityPropagation(affinity="precomputed", damping=0.9)
affprop.fit(lev_similarity)
```

```python
counter = 1
for cluster_id in np.unique(affprop.labels_):
    exemplar = words[affprop.cluster_centers_indices_[cluster_id]]
    cluster = np.unique(words[np.nonzero(affprop.labels_==cluster_id)])
    cluster_str = ", ".join(cluster)
    print("%d - *%s:* %s" % (counter, exemplar, cluster_str))
    counter += 1
```
