```python
compareData = pd.merge(GTdata, PredData, on='index')
compareData = compareData.drop(columns=['eid','original_triple_sets',
       'comment', 'lid', 'lang', 'shape',
       'shape_type', 'dbpedia_links', 'links', 'test_category'])
print(compareData.columns)
compareData = compareData.filter(['index', 'category', 'size', 'text', 'modified_triple_sets', 'prediction', 'latency'])
compareData = compareData.rename(columns={'modified_triple_sets': 'ground_truth'})
compareData.head()
```

## EVALUATION OVERALL

```python
# EVALUATION

'''
Finding similarity metric
https://stackoverflow.com/questions/17388213/find-the-similarity-metric-between-two-strings
'''

from difflib import SequenceMatcher

# s = SequenceMatcher(None, gt, pred)
# s.ratio()

debugFlag = False
debugFlag2 = False

relationsNotMatch = []

matchCount = 0
prevMatchCount = 0
actualNumOfGTtriplets = 0

noPredictions = 0
totalSentences = 0

allPrec = []
allRecall = []
allF1 = []

dispCt = 0
for ind in compareData.index:
    
    totalSentences += 1
    
    if ind < 300 and debugFlag:
        continue
    
    if dispCt > 100 and (debugFlag or debugFlag2):
        break
        
    dispCt += 1

    gt = eval(compareData['ground_truth'][ind])
    predv1 = compareData['prediction'][ind].strip()
    
    predv2 = predv1.split("\n")
    
    # REMOVE REPITITION
    predv2 = list(set(predv2))
    
    # REMOVE EMPTY ITEMS AND STOP CRITERIAS
    pred = []
    for item in predv2:
        
        if not item or item in ["Test", "Test Sentence"]:
            continue
            
        # FOR GEMMA
        elif item in ["None"]:
            continue
            
        elif "There is no" in item or "The output should be" in item or "The context does not" in item or "No relevant entities" in item:
            continue
        
        elif "|" not in item:
            if debugFlag2: print(f"Check pred - {item}")
            # continue
            
        
            
        # if "|" not in item and debugFlag2:
        #     print(f"Check pred - {item}")
            
        pred.append(item)
        
    if len(pred) == 0:
        noPredictions += 1
    
    if debugFlag: 
        print("*"*20)
        print(f"Text \t- {compareData['text'][ind]}")
        print(f"GT \t- {gt}")
        print(f"Pred \t- {pred}")
    
    actualNumOfGTtriplets += compareData['size'][ind]
    
    # if len(gt) != len(pred):
    #     continue
    
    # calculate
    for itemGT in gt:
        itemGT = itemGT.replace('"', '').lower().strip()
        GTentSplit = itemGT.split(" | ")
        
        for i, item in enumerate(GTentSplit):
            GTentSplit[i] = item.replace(" ", "_")
        
        match = {'exact_match': [], 'exact_match_wo_order': [], 'entitites_match': [], 'diff_threshold': []}
        currMatchCount = 0
        
        for itemPred in pred:
            itemPred = itemPred.strip().lower()
            PredentSplit = itemPred.split(" | ")
            removePred = 0
            currCount = 0
            
            if len(PredentSplit) != 3 or currMatchCount > 0:
                continue
                
                
            for i, item in enumerate(PredentSplit):
                PredentSplit[i] = item.replace(" ", "_").replace('"', '') #.replace("b'", "")
            
            if debugFlag:
                print(f"\nCurrent Pred \t- {PredentSplit}")
                print(f"Current GT \t- {GTentSplit}")
                # print(GTentSplit[1], PredentSplit[1], GTentSplit[1] == PredentSplit[1])
            
            # EXACT MATCH WITH ORDER
            # if itemGT == itemPred:
            if GTentSplit == PredentSplit:
                currCount += 1
                matchKey = 'exact_match'
                if debugFlag: print("EXACT MATCH WITH ORDER")
                # print(f"Exact match - {itemGT} - {itemPred}")
                
            # EXACT MATCH W/O ORDER
            if currCount == 0 and len(itemGT) == len(itemPred):
                
                if len(GTentSplit) == 3 and len(PredentSplit) == 3:
                    if GTentSplit[0] == PredentSplit[2] and GTentSplit[2] == PredentSplit[0] and GTentSplit[1] == PredentSplit[1]:
                        currCount += 1
                        matchKey = 'exact_match_wo_order'
                        if debugFlag: print("EXACT MATCH W/O ORDER")
                        # print(f"Exact match - {itemGT} - {itemPred}")
                        
            # ENTITIES MATCH AND RELATION DOESN'T
            if currCount == 0:

                if len(GTentSplit) == 3 and len(PredentSplit) == 3:
                    if ((GTentSplit[0] == PredentSplit[2] and GTentSplit[2] == PredentSplit[0])
                        or (GTentSplit[0] == PredentSplit[0] and GTentSplit[2] == PredentSplit[2])):
                        currCount += 1
                        matchKey = 'entitites_match'
                        if debugFlag: print(f"ENTITIES MATCH BUT NOT RELATION")
                        # print(f"Exact match - {itemGT} - {itemPred}")
                        
                        relationsNotMatch.append((GTentSplit[1], PredentSplit[1], compareData["text"][ind],
                                                  f"{GTentSplit[0]}-{GTentSplit[2]}",
                                                 f"{PredentSplit[0]}-{PredentSplit[2]}"))
                        
            # SENTENCE DIFF SCORE
            if currCount == 0:
                s = SequenceMatcher(None, itemGT, itemPred)
                if debugFlag: print(f"Match score - {s.ratio()}")
                if s.ratio() >= 0.9:
                    matchKey = 'diff_threshold'
                    if debugFlag: print(F"SENTENCE DIFF SCORE - Match score - {s.ratio()}")
                    currCount += 1
                        
            # matchCount += currCount
            
            if currCount > 0:
                currMatchCount += 1
                match[matchKey].append(itemPred)
                
            # break
                
        if currMatchCount > 0:
            matchCount += 1
            
        # break
    
    currPrec = 0
    if len(pred) > 0:
        currPrec = (matchCount - prevMatchCount) / len(pred)
    allPrec.append(currPrec)
    
    currRec = min(matchCount - prevMatchCount, len(gt)) / len(gt)
    allRecall.append(currRec)
    
    currF1 = 0
    if currPrec + currRec > 0:
        currF1 = 2 * currPrec * currRec / (currPrec + currRec)
    allF1.append(currF1)
    
    prevMatchCount = matchCount
    if debugFlag: print()
    # break

print(f"File = {predDataPath.split('/')[-1]}")    

print(f"\nTotal Sentences - {totalSentences}. Total GTs match - {matchCount}/{actualNumOfGTtriplets} - {matchCount/actualNumOfGTtriplets*100:.2f}")

print(f"\nAverage Precision - {sum(allPrec)/len(allPrec):.3f}")
print(f"Average Recall - {sum(allRecall)/len(allRecall):.3f}")
print(f"Average F1 - {sum(allF1)/len(allF1):.3f}")

print(f"\nNo predictions for {noPredictions}/{len(compareData['prediction'])} texts")

# Inference time
average_time = list(compareData['latency'])
average_time = sum(average_time) / len(average_time)
print(f"\nAverage Inference time - {average_time:.3f} sec")
```

```python
word1 = []
word2 = []
text = []
for idx, item in enumerate(relationsNotMatch):
    word1.append(item[0])
    word2.append(item[1])
    text.append(item[2])
    # print(idx, item)
```

```python
# !pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')

# Two lists of words
# word1 = ['country','operator','recordlabel','battle']
# word2 = ['location','operated','label','foughtat']
    
#Compute embedding for both lists
embeddings1 = model.encode(word1, convert_to_tensor=True)
embeddings2 = model.encode(word2, convert_to_tensor=True)
  
#Compute cosine-similarities
cosine_scores = util.cos_sim(embeddings1, embeddings2)
    
scores = []
for i in range(len(word1)):
    scores.append(cosine_scores[i][i])
    
#Output the pairs with their score
# for i in range(len(word1)):
#     print("Score: {:.4f} \t {} \t {}".format(cosine_scores[i][i], word1[i], word2[i]))
    
for score, w1, w2, sent in sorted(zip(scores, word1, word2, text), reverse = True):
    print("\n",sent)
    print(f"{score:.4f} \t {w1} \t {w2}")
# cosine_scores[0][0]
```
